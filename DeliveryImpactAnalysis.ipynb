```python
import pandas as pd

# Load the dataset (re-added to ensure df is defined if kernel was reset)
df = pd.read_csv('/content/customer_satisfaction_final.csv')

df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.duplicated().sum()
```

```python
df['order_id'].duplicated().sum()
```

```python
df['customer_unique_id'].duplicated().sum()
```

```python
df['review_score'].value_counts().sort_index()
```

```python
df.groupby('review_score')['delivery_time_actual'].mean()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,6))

sns.boxplot(
    x='review_score',
    y='delivery_time_actual',
    data=df
)

plt.title("Delivery Time vs Customer Satisfaction")
plt.xlabel("Review Score")
plt.ylabel("Actual Delivery Time (days)")

plt.show()
```

```python
category_review = df.groupby('product_category_name_english')['review_score'].mean().sort_values()

category_review.head(10)
```

```python
df['product_category_name_english'].value_counts().head(10)
```

```python
category_analysis = df.groupby('product_category_name_english').agg({
    'review_score':'mean',
    'order_id':'count'
})

category_analysis = category_analysis.rename(columns={
    'review_score':'avg_review',
    'order_id':'orders'
})

category_analysis.sort_values('orders', ascending=False).head(10)
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    data=category_analysis,
    x='orders',
    y='avg_review'
)

plt.title("Product Category Performance")
plt.xlabel("Number of Orders")
plt.ylabel("Average Review Score")

plt.show()
```

```python
df.groupby('review_score')['total_price'].mean()
```

```python
plt.figure(figsize=(8,6))

sns.boxplot(
    x='review_score',
    y='total_price',
    data=df
)

plt.title("Price vs Customer Satisfaction")
plt.xlabel("Review Score")
plt.ylabel("Order Price")

plt.show()
```

```python
df.groupby('payment_type')['review_score'].mean().sort_values()
```

```python
corr_matrix = df.corr(numeric_only=True)

plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")

plt.title("Correlation Matrix")
plt.show()
```
